```python
def lista_radios():
    while True:
        try:
            # solicitar al usuario ingresar una lista de números
            entrada = input("Ingresa una lista de números reales positivos (separados por coma): ")
            # convertir la entrada en una lista de números
            radios = [float(x.strip()) for x in entrada.split(",")]

            # revisar que todos los números > 0
            if all(radio > 0 for radio in radios):
                return radios
            else:
                print("Todos los números deben ser positivos")
        except ValueError: #si sale error debe decir esto y no interrumpir el programa
            print("Debes ingresar solo números reales separados por una coma")
#Calculo del area y perimetro
import math
#creacion de funcion para calcular area y perimetro juntos
def calculo(radios):
    for radio in radios:
        area = math.pi * (radio ** 2)
        perimetro = 2 * math.pi * radio
        print(f"Para un radio de {radio:.2f}, el área es {area:.2f} y el perímetro es {perimetro:.2f}.")

# aca nos da los resulyados
radios = lista_radios()
calculo(radios)
```
