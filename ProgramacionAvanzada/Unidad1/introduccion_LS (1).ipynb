```python
def lista_radios():
    while True:
        try:
            # solicitar al usuario ingresar una lista de nÃºmeros
            entrada = input("Ingresa una lista de nÃºmeros reales positivos (separados por coma): ")
            # convertir la entrada en una lista de nÃºmeros
            radios = [float(x.strip()) for x in entrada.split(",")]

            # revisar que todos los nÃºmeros > 0
            if all(radio > 0 for radio in radios):
                return radios
            else:
                print("Todos los nÃºmeros deben ser positivos")
        except ValueError: #si sale error debe decir esto y no interrumpir el programa
            print("Debes ingresar solo nÃºmeros reales separados por una coma")
#Calculo del area y perimetro
import math
#creacion de funcion para calcular area y perimetro juntos
def calculo(radios):
    for radio in radios:
        area = math.pi * (radio ** 2)
        perimetro = 2 * math.pi * radio
        print(f"Para un radio de {radio:.2f}, el Ã¡rea es {area:.2f} y el perÃ­metro es {perimetro:.2f}.")

# aca nos da los resulyados
radios = lista_radios()
calculo(radios)
```
