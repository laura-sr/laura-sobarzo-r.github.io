```python
import numpy as np
import matplotlib.pyplot as plt
import funciones as md

#  Gráfico 2D de f
x = np.linspace(-10, 10, 100) #0 y 1 son los bordes 100 la cantidad de intervalos
y_f = md.f(x)

plt.figure(figsize=(6, 8)) #numeros son pulgadas
plt.plot(x, y_f, label='$f(x) = x^2$', color='#FF0080')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.title('Gráfico 2D de $f(x) = x^2$')
plt.legend()
plt.grid()
plt.show()
#funciona bien
#---------------------------------------------------------------------------------
                                  # Gráfico de g
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z_g = md.g(x, y)

fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z_g, cmap='magma')#'viridis'
ax.set_xlabel('$x$')
ax.set_ylabel('$y$')
ax.set_zlabel('$g(x, y)$')
ax.set_title('Gráfico 3D de $g(x, y) = \sin(x) \cdot \cos(y)$')
plt.show()
```
